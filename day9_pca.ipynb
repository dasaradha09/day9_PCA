```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/Carbon Emission pca.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data['Vehicle Type'].fillna(data['Vehicle Type'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
 categorical_features=data.select_dtypes(include=["object"])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in categorical_features:
  le=LabelEncoder()
  data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=15)
```

```python
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
gb=GradientBoostingRegressor()
rf=RandomForestRegressor()
```

```python
gb.fit(x_train_pca,y_train)
rf.fit(x_train_pca,y_train)
```

```python
gb_y_pred=gb.predict(x_test_pca)
rf_y_pred=rf.predict(x_test_pca)
```

```python
print("r2 score with ",r2_score(gb_y_pred,y_test))
```

```python
 print("r2 score for ",r2_score(rf_y_pred,y_test))
```
